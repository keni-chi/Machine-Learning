```python
import pandas as pd
df = pd.read_csv('df_season.csv', index_col='Month', parse_dates=True)
df.head()
```

```python
# 移動平均の描画
```

```python
from bokeh.io import output_notebook, push_notebook, show
from bokeh.plotting import figure
from ipywidgets import interact

output_notebook()
plot = figure(plot_width=600, plot_height=300, x_axis_type='datetime')
raw = plot.line(df.index, df['season1'])
ave = plot.line(df.index, df['season1'].rolling(10).mean(), color='red')
handle = show(plot, notebook_handle=True)

@interact(n=(0, 30))
def redraw(n=10):
    ave.data_source.data = {'x': df.index, 'y': df['season1'].rolling(n).mean()}
    push_notebook(handle=handle)
```

```python
# 説明変数を選択可能とする、軸ラベル付与、凡例など追加
```

```python
from bokeh.io import output_notebook, push_notebook, show
from bokeh.plotting import figure
from ipywidgets import interact

output_notebook()
plot = figure(plot_width=600, plot_height=300, x_axis_type='datetime', title='タイトル')
raw = plot.line(
       x=df.index,            #X軸
       y=df['season1'],       #Y軸
       line_width=2,          #線の幅
       #line_color="green",    #線の色
       #line_dash="dashdot",   #線の種類（実線：solid、点線：dashed・dotted・dotdash・dashdot）
       legend_label="生値"          #凡例の表示名
      )
ave = plot.line(
       x=df.index,            #X軸
       y=df['season1'].rolling(10).mean(),       #Y軸
       line_width=2,          #線の幅
       line_color="green",    #線の色
       line_dash="dashdot",   #線の種類（実線：solid、点線：dashed・dotted・dotdash・dashdot）
       legend_label="移動平均"          #凡例の表示名
      )
handle = show(plot, notebook_handle=True)

plot.xaxis.axis_label = '年月日'
plot.yaxis.axis_label = 'value'

@interact(col=['season1', 'season2', 'season3', 'season4'], n=(0, 30))
def redraw(col='season2', n=10):
    raw.data_source.data = {'x': df.index, 'y': df[col]}
    ave.data_source.data = {'x': df.index, 'y': df[col].rolling(n).mean()}
    push_notebook(handle=handle)
```
